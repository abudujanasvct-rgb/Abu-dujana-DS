```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
# IQR Method - Remove Outliers (Only on numeric columns)
numeric_df = df.select_dtypes(include=['float64', 'int64'])
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

df_filtered = df[((numeric_df >= Q1 - 1.5 * IQR) & (numeric_df <= Q3 + 1.5 * IQR)).all(axis=1)]
print("Outliers removed! Old shape:", df.shape, "New shape:", df_filtered.shape)
df = df_filtered
```

```python
# Convert datatypes

print(df.dtypes)

```

```python
df['Hours_Studied'] = df['Hours_Studied'].astype(float)
```

```python
df
```

```python
df['Performance_Level'] = pd.cut(
    df['Exam_Score'],
    bins=[0,50,75,100],
    labels=["Low","Medium","High"]
)
```

```python
plt.figure()
plt.hist(df['Exam_Score'])
plt.title("Distribution of Exam Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

```python
# Simple Line Chart
plt.figure(figsize=(10, 5))
plt.plot(df.index, range(len(df)), marker='o', color='blue', linewidth=2)
plt.xlabel('Index')
plt.ylabel('Row Number')
plt.title('Simple Line Chart')
plt.grid(True)
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=df)
plt.title("Hours Studied vs Exam Score")
plt.show()
```

```python
plt.figure()
sns.countplot(x='Performance_Level', data=df)
plt.title("Performance Categories")
plt.show()
```

```python
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
